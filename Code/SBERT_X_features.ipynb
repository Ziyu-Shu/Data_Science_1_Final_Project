## Sentence Transformers (SBERT) 

https://sbert.net/

```python
from sentence_transformers import SentenceTransformer

# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("all_documents_cleaned.csv")
```

```python
# make sure the text column is of string type or it will cause errors
df["clean_text"] = df["clean_text"].astype(str)
df["embedding"] = df["clean_text"].apply(lambda x: model.encode(x))
```

```python
df
```

```python
# unpack the embeddings to 384 columns
embed_mat = np.vstack(df["embedding"].values)

embed_df = pd.DataFrame(
    embed_mat,
    columns=[f"emb_{i}" for i in range(embed_mat.shape[1])]
)

embed_df
```

```python
# merge embed_df with dates
df_embed = pd.concat([df["date"], embed_df], axis=1)

df_embed
```

```python
df_features = pd.read_csv("Data/document_features.csv")
```

```python
# filter df_features with has_China == 1, keep only those rows
df_features = df_features[df_features["has_china"] == 1]
```

```python
# merge df_embed and df_features on date, only keep rows that mentioned China
df_features_embed = pd.merge(df_embed, df_features, on="date", how="inner")

df_features_embed
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["week_start"] = df["date"].dt.to_period("W").apply(lambda r: r.start_time)

# Aggregate embeddings by week
weekly_embeddings = (
    df
    .groupby("week_start")["embedding"]
    .apply(lambda x: np.mean(np.stack(x), axis=0))
    .reset_index()
)

weekly_embeddings.head()
```

```python
embed_mat = np.vstack(weekly_embeddings["embedding"].values)

embed_df = pd.DataFrame(
    embed_mat,
    columns=[f"emb_{i}" for i in range(embed_mat.shape[1])]
)

embed_df
```

```python

weekly_X = pd.concat([weekly_embeddings["week_start"], embed_df], axis=1)

weekly_X.head()
```

```python
weekly_X.to_csv("weekly_X.csv", index=False)
```
